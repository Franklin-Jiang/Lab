```python
import time
import random
import matplotlib.pyplot as plt

def timeit(f):
    def i(*args,**kwargs):
        start_time=time.time()
        f(*args,**kwargs)
        end_time=time.time()
        print('Running Time: {:.4}ms'.format((end_time-start_time)*10e3))
        return (end_time-start_time)*10e3
    return i

# generating random number
def generateLst(number:int):
    return [random.randint(-10e5,10e5) for _ in range(number)]


# write bubble sort function
# @timeit

def bubble(lst:list):
    compare_times,swap_time=0,0
    for i in range(len(lst)):
        for j in range(len(lst)-1,i,-1):
            compare_times+=1
            if lst[j-1]>lst[j]:
                lst[j-1],lst[j]=lst[j],lst[j-1]
                swap_time+=1
    return compare_times,swap_time
```

```python
n=[]
compare_times_lst=[]
swap_times_lst=[]
for i in range(10,301,10):
    n.append(i)
    compare_times,swap_times=bubble(generateLst(i))
    compare_times_lst.append(compare_times)
    swap_times_lst.append(swap_times)

plt.plot(n,compare_times_lst,marker='o')
```
